```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
#let's create the vectorizer and fit the corpus and transform them accordingly
v = TfidfVectorizer()
v.fit(corpus)
transform_output = v.transform(corpus)
print(transform_output) # i don't know what this does
```

```python
#let's print the vocabulary

print(v.vocabulary_)
```

```python
all_feature_names = v.get_feature_names_out()
# all_feature_names is basically a bags of all the words present in our entire dataset in alphabetical order
for word in all_feature_names:
  indx = v.vocabulary_.get(word) # basically each word ko index get gareko xa, although alphabetical order ma thiyo so i don't know sidhai index use gare huntyo enumeratorko
  idf_score = v.idf_[indx] # basically tyo word ko idf kati ho nikaleko ho
  print(f"{word} : {idf_score}")
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/ecommerceDataset.csv" , header=None)
df.columns = ['label', 'text']
```

```python
df.info()
```

```python
df['label'].value_counts()
```

```python
df['label_num'] = df['label'].map({
    'Household' : 0,
    'Books': 1,
    'Electronics': 2,
    'Clothing & Accessories': 3
})
df.info()
```

```python
df.tail(5)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
df.dropna(inplace=True)
df.info()
X_train, X_test, y_train, y_test = train_test_split(
    df.text,
    df.label_num,
    test_size=0.2, # 20% samples will go to test dataset
    random_state=2022,
    stratify=df.label_num
) # here the equal distribution is maintained on the basis of hte df.label_num.

# as you can see
print(y_train.value_counts() / len(y_train))
print(y_test.value_counts() / len(y_test))
# you can see they have the same distribution.
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()), # this is transformer
     ('KNN', KNeighborsClassifier())   # this one is estimator.
])
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),
     ('Multi NB', MultinomialNB())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),        #using the ngram_range parameter
     ('Random Forest', RandomForestClassifier())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

### since this random forest classifier had the best weighted avg f1-score this is actually preferrred.

```python
import spacy
nlp = spacy.load('en_core_web_sm') # basically we are trying to lemmatize the word and remove stop words and punctuations. so since these factors depend mostly on language like english has different stop words , punctuations and lemmatized variation what language must be specified

# for example

val = nlp('I am having dinner right now! , it is the best dinner i have had in years ')
print(type(val)) # type of spacy token
for token in val:
  if token.is_stop or token.is_punct:
    continue
  print(token.lemma_ , end=" ")
```

```python

def preprocess(text):
    # remove stop words and lemmatize the text
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
df['preprocessed_txt'] = df['text'].apply(preprocess)
```

```python
df.head(5)
```

```python
X_train , X_test , y_train, y_test  = train_test_split(
    df.preprocessed_txt,
    df.label_num,
    test_size = 0.2,
    random_state=2022,
    stratify=df.label_num
)
```

```python
#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),        #using the ngram_range parameter
     ('Random Forest', RandomForestClassifier())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

### you can see how removing the stop words, punctuating, and lemmatizing improves the weighted avg of f1-score to 0.97 from 0.96
